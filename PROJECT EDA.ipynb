## scatter plot: SHOWS THE RELATIONSHIP BETWEEN SALARY AND AGE

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x='Age',y='Salary',color='teal',s=100,hue='Gender')

plt.title('relationship between salry and age')
plt.xlabel('ages')
plt.ylabel('salary')

plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## BOX PLOT and VOILINPLOT: DEPICT THE RELATIONSHIP BETWEEN GENDER AND SALARY

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df,x='Gender',y='Salary',palette='Set3')

plt.title('Relationship between gender and salary')
plt.xlabel('gender')
plt.ylabel('SALARY')


plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.violinplot(data=df,x='Gender',y='Salary',palette='Set2')

plt.title('violinplot: salary by country')
plt.xlabel('gender')
plt.ylabel('SALARY')


plt.grid(True)
plt.tight_layout()
plt.show()
```

## box plot and scatter plot : ILLUSTRATE the connection between education level and salary

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df,x='Education Level',y='Salary',palette='Set3')

plt.title('box plot: Eduaction and salary')
plt.xlabel('education level')
plt.ylabel('SALARY')


plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.violinplot(data=df,x='Education Level',y='Salary',palette='Set3')

plt.title('violin plot: Eduaction vs salary')
plt.xlabel('education level')
plt.ylabel('SALARY')


plt.grid(True)
plt.tight_layout()
plt.show()
```

## bar plot : avg salary for different job titles

```python
# step 1calculate avg salary per job title
avg_salary = df.groupby('Job Title')['Salary'].mean().reset_index()


#step2  sort avg salary by for  a cleaner data plo
avg_salary = avg_salary.sort_values(by = 'Salary',ascending=False)

#step 3 plot 

plt.figure(figsize=(10,5))
sns.barplot(data=avg_salary,x='Job Title',y='Salary',palette='Blues_d')

plt.title('AVG SALARY BY JOB TITLE ')
plt.xlabel('JOB TITLE')
plt.ylabel('SALARY')


plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# ---------- 6. Heatmap: Correlation Matrix ----------
numeric_cols = df.select_dtypes(include=['float64', 'int64'])
corr_matrix = numeric_cols.corr()

plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
# ---------- 7. Pair Plot: Numerical Columns ----------
sns.pairplot(numeric_cols)
plt.suptitle('Pair Plot of Numerical Features', y=1.03)
plt.show()
```
